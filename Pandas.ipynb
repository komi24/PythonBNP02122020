```python
df = pd.read_csv("train.csv")

df = df.drop(columns=["Cabin"])
df["Sex"] = df["Sex"].map({"female": 1, "male": 0})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})
df
```

```python
df["Embarked"].unique()
```

```python
import re

def convert_title(name):
    return re.findall(", ([^.]*)\.", name)[0]

df["Title"] = df["Name"].map(convert_title) # on passe la fonction convert_title et non le resultat de convert_title("toto")
df
```

```python
df["Title"].unique()
```

```python
df["Age x 2"] = df["Age"] * 2
df
```

```python
def convert_age(age):
    return age * 2

df["Age * 2"] = df["Age"].map(convert_age) # on passe la fonction convert_title et non le resultat de convert_title("toto")
df
```

```python
df["Age"].mean()
df["Age"].sum()

# Moyenne d'age des survivants
# Pourcentage des hommes morts par rapport au nombre d'hommes
# Pourcentage des femmes mortes par rapport au nombre de femmes
print(df[df["Survived"] == 1]["Age"].mean())
print(len(df[(df["Survived"] == 0) & (df["Sex"] == 0)]) / len(df[df["Sex"] == 0]))
print(len(df[(df["Survived"] == 0) & (df["Sex"] == 1)]) / len(df[df["Sex"] == 1]))
```
