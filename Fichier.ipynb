```python
import requests

res = requests.get("https://raw.githubusercontent.com/komi24/PythonBNP02122020/master/generated.json")

# res.content
with open("gen.json", "wb") as f:
    f.write(res.content)
```

```python
import json

liste_personnes = []
with open("gen.json", "r", encoding="utf-8") as f:
    liste_personnes = json.loads(f.read())
    
liste_personnes[0]
```

```python
len(liste_personnes)
```

```python
with open("mail_template.txt", "w", encoding="utf-8") as f:
    f.write("Email : {{email}},\n"+
        "Bonjour {{name}}, Vous tes intress par {{categorie}}.\n"+
        "Nous avons un produit qui pourrait vous intresser.\n"+
        "Prix : {{price}}\n")
```

```python
# Pour tous les elements de liste_personnes ajouter au fichier "mailing.txt",
# le template rempli par les donnes correspondantes

with open("mail_template.txt", encoding="utf-8") as f:
    template = f.read()

with open("mailing.txt", "w", encoding="utf-8") as f:
    for personne in liste_personnes:
        f.write(
            template
                .replace("{{email}}", personne['email'])
                .replace("{{name}}", personne['name'])
                .replace("{{categorie}}", personne['categorie'])
                .replace("{{price}}", str(personne['price']))
        )
    
```
